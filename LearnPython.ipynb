# STDOUT / PRINT
---
Print is the way python displays message/text on the CLI where users can read.<br>
*(Also sometimes developers like to use stdout rather than using specialized debugging tools)*<br><br>

## PRINT ARGUMENTS
---
- `*objects`: text to display. Can be inputted as much as you want with comma as separator.
- `end`: ends with (default to `'\n'`).
- `sep`: separate each string with, usually with comma (default to `' '`/space).
- `flush`: force to print even if still executing code (def: `False`).
- `file`: where output will be written (def: `sys.stdout`/CLI Out).

(Other than `*objects`, all is optional. to set their value uses like `print("...", end="!")` etc.)

````python
# stdout
print("Hello, World!") # default
print("Hell" "o") # string-slice (like how char behaves) (OUT: str 1 + str 2 ...)
print("Hello", "Brian") # comma separated (OUT: str 1 + SPACE + str 2 ...)
print("Good" + "bye") # string combination (same OUT as string slice)

print('\n') # Extra line break for clean output

# Format
print(f"i'm {2025-2008} years old") # f-string, formattable using {} with var, calculation, format, etc. to display
print("price is {:.2f} dollars".format(45)) # .format() func, uses {} too
print("my brother name was %s" % "Rachey".upper()) # % formatting (C-based like java too). %s string, %d integer, etc. If multiple arguments, use print(..str.. % (..format-separated-by-comma..))
````

# VARIABLES
---
Variable is data where you save the value to desired name. Best advice is using `camelCase` or use underscore (_) separated each word for writing var name, and also use 'meaningful' name for that. Also must be remembered that naming var ***CANNOT*** be the name of python valid defined keyword or function, such as def, with, as, open, etc.
<br><br>
## VAR TYPES
---
- `str` (string) = basically a text. Uses qoutation marks at start and end `' '` or `" "`. In C, str is just char-list which also valid in python.
- `int` (integer) = round number with the range of `-2,147,483,648` until `2,147,483,647` (32-bit). Write it without comma or any other separators, otherwise invalid. If you struggled to read big value, you can use underscore as separator.
- `float` = decimal (separated by peridot '`.`' not comma).
- `bool` (boolean) = condition. `True` / `False` (must be Capitalized at start, unlike other languages).
- `list` (array) = mutable data list, separated by comma. Stored in `[...]`.
- `dict` (dictionary) = set of data with its value, separated by comma. Stored in `{'key': value}`.
- `tuple` = immutable data list, separated by comma. stored in `(...)`.
- `None` = '`undefined`'-value version of python.

````python
# Variables
name = "Frederick" # str (string)
child = 2 # int (integer)
age = 42.8 # float
has_job = True # bool (boolean)
friends = ["Allie","Herbert","Corgi"] # list (array)
corporationData = { # dict (dictionary)
  "name": "Golden Bolt",
  "salary": 1260,
  "isLegal": True
}
familyParts = ( # tuple
  "wife",
  "child",
  "child",
  "dad",
  "mom",
  "gradma"
)
friendWorkSameShift = None # None (undefined)
print(
  "My name is", name,
  "\nI have", child, "child",
  f"\nI {'have' if has_job else 'dont have'} a job",
  f"\nMy friends are {', '.join(friends)}",
  f"\nMy corporation was {corporationData['name']}",
  f"\nMy family parts was: {', '.join(familyParts)}",
  f"\n{friendWorkSameShift} of my friends work same shift as me."
)
````

# CONDITION
---
This consist of `if`, `elif`, `eLse`, and `match`-`case`.
- `if` is used to check whenever condition is True, it'll run code-block or sector below it.
- `elif` (else If) is used to check next step condition when If was unmet condition.
- `else` used when everything has passed resulting in False condition. This doesn't need extra condition input.
- `match`-`case` is chaining condition checking just like `if`-`elif`-`else`, but it just includes 1 input at `match` to be checked by each `case`.

<br><br>
### EXAMPLE
---
> `if`-`elif`-`else`

```python
if <CONDITION>:
  <CODE>
  ...
elif <CONDITION>:
  <CODE>
  ...
else:
  <CODE>
  ...
```
> `match`-`case`

```python
match <INPUT>:
  case <VALUE-1>:
    <CODE>
    ...
  case <VAlUE-2>:
    <CODE>
    ...
  case _: # This works like `else`
    <CODE>
    ...
```
<br><br>
## LOGIC
---
Logic is where you can modify the condition, like mathematical operation but for boolean. It's based on 'Logic Gates'.
- `and` is used to check both value are `True`.
- `or` is used to check at least one value `True`.
- `not` is used to reverse condition. Used at the beginning.

> NOTE: Logic priority can be changed using `(...)`.

<br><br>
### EXAMPLE
---

```python
# AND
True and True # True
True and False # False
False and False # False

# OR
True or True # True
True or False # True
False or False # False

# NOT
not True # False
not False # True

# Priority Changing
not (True or False) and not False # False
```


````python
# Condition
thereIs = {
  "Anna": True,
  "Hana": True,
  "Zoeyd": False,
}

# Include to dictionary
thereIs["Benn"] = True
thereIs["Yarren"] = False

# If only
if thereIs["Anna"]:
  print("There is Anna")

print('\n')

# If-Else
if thereIs["Benn"]:
  print("There is Benn")
else:
  print("Benn is absent")

print('\n')

# If-Elif-Else + and + or
if thereIs["Yarren"] and thereIs["Hana"]:
  print("Both Yarren and Hana is here")
elif thereIs["Yarren"] or thereIs["Hana"]:
  print("One of Yarren or Hana is here")
else:
  print("Both Yarren and Hana aren't here")

print('\n')

# If-Not
if not thereIs["Zoeyd"]:
  print("Zoeyd isn't here")
else:
  print("Zoeyd here")

print('\n')

account = {
  "Anna":"annaleyrand@mail.com",
  "Hana":"hanabill@bill.go",
  "Zoeyd":"zoey.curious@mail.com",
  "Benn":"benny.rachero622@epsilon.academy",
  "Yarren":"yarrenharrysharg@mail.com"
}

emailTo = account["Benn"]

# Match case
match emailTo:
  case "annaleyrand@mail.com":
    print("Contacting Anna")
  case "hanabill@bill.go":
    print("Contacting Hana")
  case "zoey.curious@mail.com":
    print("Contacting Zoeyd")
  case "benny.rachero622@epsilon.academy":
    print("Contacting Benn")
  case "yarrenharrysharg@mail.com":
    print("Contacting Yarren")
  case _:
    print("No contact found!")
````

# LOOPING
---
Code looping can be achieved by using `for` or `while` keyword.
- `for` is used to iterates/loops code for desired amount of loop.
- `while` is used to loop when the condition is `True`, just like `if`.

<br><br>
## EXAMPLE
---
```python
# for in range
for <ASSIGNED_VAR> in range(START,STOP,STEP):
  <CODE>
  ...

# for array/list iteration
for <ASSIGNED_VAR> in <ARRAY>:
  <CODE>
  ...

# for dict iteration
for <ASSIGNED_KEY>, <ASSIGNED_VALUE> in <DICT>.items():
  <CODE>
  ...

# while
while <CONDITION>:
  <CODE>
  ...
```

````python
# Loop
day = ["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]
grade = {
  "Anna": 80,
  "Gary": 78,
  "Verion": 85
}

# base loop for
for i in range(8): # iterates 8 times in 0-7 range
  print(f"Iteration: {i}", flush=True)

print('\n')

# array iter
for i in day:
  print(f"Day: {i}", flush=True)

print('\n')

# dict iter
for k,v in grade.items():
  print(f"Name: {k.upper()} - Grade: {v}", flush=True)

print('\n')

# while
delivery = 0
while delivery < 7:
  print(f"Delivering ID-{delivery}", flush=True)
  delivery += 1 # Increment to limit, if u forgot = ur device bricked
````

# ERROR HANDLING
---
Error handling is alternative code to executes when error happened. It checks for code in `try` and then executes `except` when error. There also `finally` key which just few people used it. It **ALWAYS** executed no matter it's error or not, so almost nobody used it (optional).
<br><br>
## EXAMPLE
---
```python
# accept all error types
try:
  <CODE>
  ...
except:
  <CODE>
  ...
finally:
  <CODE>
  ...

# filtering
...
except <ERROR_CLASS>:
  <CODE>
  ...

# multi-filter single line, for differ errors with same execution
...
except (<ERROR_CLASS_1>, <ERROR_CLASS_2>, ...):
  <CODE>
  ...

# multi-filtering ladder, for differ errors with individual execution
...
except <ERR_1>:
  <CODE>
  ...
except <ERR_2>:
  <CODE>
  ...
...

# error alias (error message saved to variable)
...
except <ERR> as <ASSIGNED_VAR>:
  ...
```

````python
# Error handling
unsorted = ["Harold", 82, 39, 58, "Anna", "Rachel", 80]

# Strict Try-Except
for i in unsorted:
  try:
    print(i-12, flush=True)
  except TypeError:
    print("This isn't valid number", flush=True)

print('\n')

# Free Try-Except
for i in unsorted:
  try:
    print(i.upper(), flush=True)
  except:
    print("Not a string", flush=True)

print('\n')

# Alias Try-Except
divp = [100, "X", 60, 26, "Y", 0]
for i in divp:
  try:
    print(100/i,flush=True)
  except TypeError as e: # Alias
    print(f"Error type can't be divided: {type(i)}\n-INFO: {e}", flush=True)
  except ZeroDivisionError:
    print("Zero division is invalid", flush=True)
````

# FUNCTION
---
Function is a way to define your own mechanism or compacted code that can accept input and do execution inside itself. Python uses `def` key to define function.
<br><br>
## EXAMPLE
---
```python
# no-argument func
def <FUNCTION_NAME>():
  <CODE>
  ...

# calling
<FUNCTION_NAME>()

### EXAMPLE:
def say_hello():
  print("hello!")

say_hello()

#######

# with arguments
def <NAME>(<PARAMETER>):
  <CODE>
  ...

# calling
<FUNC>(<ARGUMENT>)

### EX:
def loop_hello(amount):
  for i in range(amount):
    print("hello!")

loop_hello(6)

#######

# value returned function
def <NAME>(...):
  ...
  return <VALUE>

# this assigned to variable
<VARIABLE> = <NAME>(...)

### EX:
def mutiply(a,b):
  return a*b

# not doing anything
multiply(3,4)

# assign to var and print
result = multiply(3,4)
print(result)
```

````python
# Function
def call(name):
  print(f"Hello {name}")

def square(num):
  return num**2 # return will not print, but can be used to assign value to var

call("Harold")
quad = square(4)
print(quad)
````

# OOP
---
OOP (Object Oriented Programming) or you can say `class` programming, is way to define your own object variable type. Like other var types, `str`, `int`, etc. they just a class. `class` has many aspect to learn, but I'll just teach you `__init__` and class-function.<br>
- `class` is object where it has many usage. In simple term, it's object definition or variable type. The best case to name `class` is using `CapitalizedCase` which capitalized each words.
- `__init__` is identity of `class` structure that can saves variables, code, etc. for the class itself, defined by `def __init__(self, ...): <...>`. `__init__` is one of *magic methods* or *dunder function (double-underscore)*. Class has many dunder, but not all must be learned for the first time.
- *'class-function'* refer to function defined inside `class` itself. Mostly it needs its `class` data, but there also class-function that freely usable without the connection of its `class`, called *'staticmethod'* which can be defined by `@staticmethod` decorator. But as I said, we will just learn class-function.

<br><br>
## CODE SNIPPET
---
```python
# class definition
class <CLASS_NAME>:
  
  # something like variables
  ...

  # init
  def __init__(self, <PARAMETER>, ...):
    ...
  
  # another things
  ...

# assign class to variable
myNewClass = <CLASS>(<INIT_ARGUMENTS>)
```

````python
# OOP
class Person:
  def __init__(self, name, age, education):
    self.name = name
    self.age = age
    self.education = education

  def introduce(self): # use self to fetch class value
    print(f"Hi, I'm {self.name}. I'm {self.age} years old. My education strength was {self.education}.\n")

# multiple var assignment, separated by comma [A,B,C = A,B,C]
hana, william, zaiko = Person("Hana", 28, "Good"), Person("William", 34, "Average"), Person("Zaiko", 30, "Mid")

hana.introduce() # `.introduce()` function calling
william.introduce() # it needs its class attachment `.FUNCTION_NAME()`
zaiko.introduce() # can be direct calling by `CLASS.FUNC()` or variable assigned `VAR.FUNC()`
````

# LIBRARY
---
Library/Lib is package or external module which can be used after `import`-ed. `import` is a key to get the lib. There also pairs with `import`, the `from` and `as`.<br>
There are many pre-installed library as you installed the interpreter, such as: `os` (OS mechanism), `math` (Mathematical operation), `json` (JSON utility), etc.<br>
For non-preinstalled lib, you can use `pip install <LIB_NAME>` on the CLI to install the library locally.
<br><br>
## EXAMPLE
---
> Importing

```python
# import only
import <LIB>

# import specified
from <LIB> import <SOMETHING>, ...

# import everything without library name attachment
from <LIB> import *

# import with alias
import <LIB> as <CUSTOM_NAME>

# import specified with alias
from <LIB> import <THING> as <NAME>
```
> Library calling

```python
# 1st line of import
<LIBRARY>.<THING>

# 2nd
<SOMETHING>

# 3rd
<LIBRARY_THING>

# 4th
<CUSTOM_NAME_LIB>.<SOMETHING>

# 5th
<CUSTOM_DEFINED>
```

````python
# must be on top
# math + random
import math
import random as rd

### math
# sqrt (square root). Achievable using x**0.5 without math lib
print("9: ",math.sqrt(9))
print("36: ",math.sqrt(36))
print("81: ",math.sqrt(81))
print("100: ",math.sqrt(100))

print('\n')

# sin. Measured in radians
print("30: ",math.sin(30))
print("60: ",math.sin(60))
print("90: ",math.sin(90))
print("0: ",math.sin(0))

print('\n')

# power. Achievable by x**y w/o math
print("3^2: ",math.pow(3,2))
print("3^3: ",math.pow(3,3))
print("4^2: ",math.pow(4,2))

print('\n')

# modulo. Remainder of division, achievable by x % y w/o math, but may differ
print("5 % 2: ",math.fmod(5,2))
print("7 % 3: ",math.fmod(7,3))
print("17 % 5: ",math.fmod(17,5))

print('\n')

### random
# random (0-1)
def generator(amt:float=10,loop:int=5)->None:
  for i in range(loop):
    print(f"Random val from {amt} = {rd.random()*amt:.2f}")

generator()
generator(68,3)

print('\n')

# choice. Randomly select one value out of array, poll equally
def gacha(loop:int=5)->list:
  data = ["Prize A", "Prize B", "Prize C", "None", "Surprise A", "Surprise B", "Surprise C"]
  gathered = [rd.choice(data) for x in range(loop)] # list comprehension, just wait
  return gathered

print(gacha())
print(gacha(3))
````

````python
### Extra
## Type annotation
# In variable
text: str = "Hello, I'm not a monster"
price: int = 12
clear_weather: bool = "Mid" # Doesn't bother, still valid
print(text,price,clear_weather, sep="\n")

print('\n')

# In function
def hello(name: str) -> str:
  return f"Hello {name}"

print(hello("Zorris"))

print('\n')

## Single line if-else/Ternary
num = [3,8,3,6,0,4,8,4,1,2,4,1,8,9]
odd = []
for x in num:
  odd.append(x if x%2!=0 else "next")

print(odd)

print('\n')

# List comprehension
even = [x for x in num if x%2==0]
print(even)

print('\n')

# Lambda
add = lambda x,y: x+y
print(add(9,5))

print('\n')

# STDIN
name = input("What's your name? ")
print(f"Hello {name}")

print('\n')

# dunder
class Animal:
  def __init__(self, name:str, color:str):
    self.name = name
    self.color = color

  def __repr__(self):
    return f"Animal name: {self.name}\nAnimal color: {self.color}"

  def __str__(self):
    return self.name

  def __eq__(self,compare): # comparing color
    if not isinstance(compare,Animal):
      return False
    return self.color == compare.color

cat_o = Animal("cat","orange")
dog_b = Animal("dog","black")
cat_w = Animal("cat","white")
dog_o = Animal("dog","orange")

print(repr(cat_o)) # call repr
print(str(cat_o),"\n") # call str
print(repr(dog_b))
print(str(dog_b),"\n")

print(cat_w == dog_b) # False
print(cat_o == cat_w) # False
print(cat_o == dog_o,"\n") # True
````
